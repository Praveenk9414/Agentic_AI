### 1. Autonomy

    Autonomy refers to the AI system's ability to make decisions and take actions on its own to achieve a given goal, without needing step-by-step human instructions.

Lets take an example of An AI recruiter application ->

1. Our AI recruter is autonomous
2. It's proactive ... whereas GenAI is reactive
3. Autonomy in multiple facets
    a. Execution -> executes each steps automatically
    b. Decision making -> shortlisting candidates automatically
    c. Tool usage -> have various tools to perform these above operations

4. Autonomy can be controlled
    1. Permission Scope - Limit what tools or actions the agent can perform independently. (Can screen candidates, but needs approval before rejecting anyone.)
    2. Human-in-the-Loop (HITL) - Insert checkpoints where human approval is required before continuing. (Can I post this JD)
    3. Override Controls - Allow users to stop, pause, or change the agent's behaviour at any time. (pause screening command to halt resume processing.)
    4. Guardrails / Policies - Define hard rules or ethical boundaries the agent must follow. (Never schedule interviews on weekends)
6. Autonomy can be dangerous
    1. The application autonomously sends out job offers with incorrect salaries or terms.
    2. The application shortlists candidates by age or nationality, violating anti-discrimination laws.
    3. The applications spending extra on LinkedIn ads.
       
   And this is why autonomy is being controlled [4]

### 2. Goal Oriented

Being goal-oriented means that the AI system operates with a persistent objective in mind and continuously directs its actions to achieve that objective, rather than just responding to isolated prompts.

    1. Goals acts as a compass for Autonomy 
    2. Goals can come with constraints 
    3. Goals are stored in core memory 
    4. Goals can be altered (for e.g; due to low application_received ... the HR might change the goal from Backend Engg to Freelancer)


### 3. Planning
Planning is the agent's ability to break down a high-level goal into a structured sequence of actions or subgoals and decide the best path to achieve the desired outcome.

Step 1: Generating multiple candidate plans

    - Plan A: Post JD on Linkedin, GitHub Jobs, AngelList 
    - Plan B: Use internal referrals and hiring agencies
Step 2: Evaluate each plan

    • Efficiency (Which is faster?)
    • Tool Availability (Which tools are available) -> for e.g; if you have to follow plan-A but donot have Google search api so you switch to alternative
    • Cost (Does it require premium tools?)
    • Risk (Will it fail if we get no applicants?)
    • Alignment with constraints (remote-only? budget?)
Step 3: Select the best plan with the help of:

    • Human-in-the-loop input (e.g., "Which of these options do you prefer?")
    • A pre-programmed policy (e.g., "Favor low-cost channels first")

### 4. Reasoning

Reasoning is the cognitive process through which an agentic ai system interprets information, draws conclusions, and makes decisions - both while planning ahead and while executing actions in real time.
Reasoning During Planning:

1. Goal decomposition - Break down abstract goals into concrete steps
2. Tool selection - Decide which tools will be needed for which steps
3. Resource estimation - Estimate time, dependencies, risks
   
Reasoning During Execution:

1. Decision-making - Choosing between options (3 candidates match → schedule 2 best, reject 1)
2. HITL handling - Knowing when to pause and ask for help (Unsure about salary range)
3. Error handling - Interpreting tool/API failures and recovering

### 5. Adaptability
Adaptability is the agent's ability to modify its plans, strategies, or actions in response to unexpected conditions — all while staying aligned with the goal.

1. Failures (failed to fetch Calendar API)
2. External feedback (Less no of applications)
3. Changing goals (Hiring a freelancer)

### 6. Context Awareness
Context awareness is the agent's ability to understand, retain, and utilize relevant information from the ongoing task, past interactions, user preferences, and environmental cues to make better decisions throughout a multi-step process.
(b/c this multistep process can sometime take several days, weeks or even months ... and so that when resumed it has the context)

1. Types of context
    1. The original goal ... (keep track of the og goal)
    2. Progress till now + Interaction history ... (Job description was finalized & posted to LinkedIn & GitHub Jobs)    
    3. Environment state ... (Number of applicants so far = 8 (or) LinkedIn promotion ends in 2 days)
    4. Tool responses ... (Resume parser → "Candidate B has 3 years Django + AWS experience or Calendar API → "No conflicts at 2 PM Wednesday)
    5. User specific preferences ...(Prefers remote-first candidates (or) Likes receiving interview questions in a Google Doc)
    6. Policy or Guardrails ... (Do not send offer without explicit user approval or Never use platforms that require paid ads unless approved)

2. Context awareness is implemented through memory ... (i.e all the above context is stored in the memory)
    1. Short term memory (e.g candidate info)
    2. Long term memory (e.g Not to send offer letter w/o approval)
